```python
file = open("data.text","a")
file.write("\nWelcome to the file handling")
file.close()
```

```python
file=open("data.text","r")
content=file.read()
sum=1
for i in content:
    if i =="\n":
        sum+=1
print(content)
print(sum)
file.close()
```

```python
file=open("data.text","r")
content=file.read()
lines= content.splitlines()
print(len(lines))
file.close()
```

```python
file=open("data.text","r")
content=file.read()
count=0
for _ in content:
  count+=1
print(count)
```
